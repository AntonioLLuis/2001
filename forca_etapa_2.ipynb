```python
Etapa 2 — Escolher palavra aleatória de uma lista (sem acentos)

Agora vamos ter uma lista e escolher aleatoriamente uma palavra.

# forca_lista.py
import random

WORD_LIST = ["python", "computador", "algoritmo", "projeto", "arquivo"]

palavra = random.choice(WORD_LIST)
letras_reveladas = ["_"] * len(palavra)

while True:
    print("Palavra:", " ".join(letras_reveladas))
    chute = input("Digite uma letra: ").strip().lower()
    if not chute:
        print("Digite algo.")
        continue
    letra = chute[0]
    for i, ch in enumerate(palavra):
        if ch == letra:
            letras_reveladas[i] = ch
    if "_" not in letras_reveladas:
        print("Você venceu! Palavra:", palavra)
        break

O que mudou

import random e random.choice(WORD_LIST) — escolhe uma palavra aleatória da lista.

A lista WORD_LIST contém várias palavras sem acento.

Teste: rode múltiplas vezes para ver diferentes palavras.
```
